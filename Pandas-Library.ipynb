```python
# combine two series
```

```python
s1 = pd.Series([1,2],index = ['a','b'], name = 's1')
s2 = pd.Series([3,4],index = ['a','b'], name = 's2')
```

```python
s1
```

```python
s2
```

```python
# concat  = combine multiple series to form dataframe

row_wise = pd.concat([s1,s2], axis = 0)
row_wise
```

```python
type(row_wise)
```

```python
row_wise.shape
```

```python
column_wise = pd.concat([s1,s2], axis = 1)
column_wise
```

```python
type(column_wise)
```
