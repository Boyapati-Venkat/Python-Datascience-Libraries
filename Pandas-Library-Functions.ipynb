```python
# merge
```

```python
stu_details = {'id':[1,2,3,4,5],
               'name':['P','Q','R','S','T'],
               'region':['N','S','W','W','S']}

stu_marks = {'id':[1,2,3,4,6],
             'sub':['Py','matlab','R','Py','Py'],
             'marks':[10,13,14,16,13]}
```

```python
df1 = pd.DataFrame(stu_details)
df2 = pd.DataFrame(stu_marks)
```

```python
df1
```

```python
df2
```

```python
# inner merge = common between both the tables
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='id')
```

```python
# left merge : fetch everything from left table and common one from right table
pd.merge(df1,df2,on = 'id',how = 'left')
```

```python
# right merge : fetch everything from right table and common one from left table
pd.merge(df1,df2,on = 'id',how = 'right')
```

```python
pd.merge(df1,df2,how = 'right')
```

```python
# operation on merged data
pd.merge(df1,df2,how = 'right').loc[:,['region','sub','marks']]
```

```python
# outer merge : union of both the tables
pd.merge(df1,df2,on = 'id',how = 'outer')
```

```python
# merge data when column names are not similar
```

```python
stu_details = {'s_id':[1,2,3,4,5],
               'name':['P','Q','R','S','T'],
               'region':['N','S','W','W','S']}

stu_marks = {'id':[1,2,3,4,6],
             'sub':['Py','matlab','R','Py','Py'],
             'marks':[10,13,14,16,13]}

df3 = pd.DataFrame(stu_details)
df4 = pd.DataFrame(stu_marks)
```

```python
df3
```

```python
df4
```

```python
# inner merge
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,left_on='s_id',right_on='id',how = 'inner')
```

```python
pd.merge(df3,df4,left_on='s_id',right_on='id',how = 'left')
```
